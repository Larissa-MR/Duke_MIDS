### Excercise 3

Now check this relationship using statsmodels. Regress birthweight on gestational period and whether the infant’s mother smoked.


---

```python
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf

result = smf.ols('bwt_oz ~ gestation + smoke', data = gest_225).fit().summary()
result
```

### Excercise 4

Now let’s expand our model to also take into account mothers’ pregnancy weight and race (make sure to treat race as a categorical variable! If you’re rusty on categoricals and indicator variables, here’s a little refresher.).

---

```python
result1 = smf.ols('bwt_oz ~ gestation + smoke + mpregwt + C(mrace)', data = gest_225).fit()
result1.summary()
```

### Excercise 5

Now let’s test for whether there is an interaction between the mother’s race and the effect of smoking.

---

```python
num_values = [0,1,2,3,4,5]
gest_225['race'] = gest_225.mrace.apply(lambda x : 'White' if x in num_values else 'Non-white')
```

```python
gest_225.head()
```

```python
result3 = smf.ols('bwt_oz ~ C(race)*smoke ', data = gest_225).fit()
result3.summary()
```
